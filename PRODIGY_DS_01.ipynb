```python
# width = 20 and height = 5
plt.figure(figsize=(20, 5))
# x - categorical
# y - continuous

ax = sns.barplot(x = 'Brands',
            y = 'Rating', ci = None, 
            data = data)
ax.set(xlabel="Brands", ylabel = "Ratings")
plt.title('Ratings across Brands')
plt.show()
```
